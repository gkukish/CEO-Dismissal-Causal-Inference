```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('final_dataset_mine.csv')
df.info()
```

```python
df.hist(bins=21, figsize=(20,15))
```

```python
from tabulate import tabulate
df.describe(include='all')
# Calculate summary statistics
summary_stats = df.describe(include='all')

# Convert summary statistics to LaTeX table format
latex_table = tabulate(summary_stats, headers='keys', tablefmt='latex')

# Save the LaTeX table to a file
with open('summary_stats.tex', 'w') as f:
    f.write(latex_table)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(), annot=True, cmap='YlGnBu')
```
